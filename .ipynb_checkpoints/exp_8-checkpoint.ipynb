```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

# Generate a larger random dataset
np.random.seed(42)
x_data = np.linspace(0, 10, 100)
y_data = 2 * x_data + 1 + np.random.normal(0, 2, 100)

# Convert to TensorFlow tensors
x_data_tf = tf.constant(x_data, dtype=tf.float32)
y_data_tf = tf.constant(y_data, dtype=tf.float32)

# Initialize variables for slope (W) and intercept (b)
W = tf.Variable(tf.random.normal([1], mean=0.0, stddev=1.0), name="weight")
b = tf.Variable(tf.zeros([1]), name="bias")

# Set the learning rate and number of epochs
learning_rate = 0.01
epochs = 100

# Define the linear regression model
def linear_regression(x):
    return W * x + b

# Define the mean squared error (MSE) loss function
def mean_squared_error(y_pred, y_true):
    return tf.reduce_mean(tf.square(y_pred - y_true))

# Use the Stochastic Gradient Descent (SGD) optimizer
optimizer = tf.optimizers.SGD(learning_rate)

# Lists to store the history of weights, biases, and losses
weights_history = []
biases_history = []
loss_history = []

# Training loop
for epoch in range(epochs):
    with tf.GradientTape() as tape:
        # Compute the predicted y values
        y_pred = linear_regression(x_data_tf)

        # Compute the mean squared error
        loss = mean_squared_error(y_pred, y_data_tf)

    # Calculate gradients with respect to W and b
    gradients = tape.gradient(loss, [W, b])

    # Update weights and biases using the optimizer
    optimizer.apply_gradients(zip(gradients, [W, b]))

    # Store the history for plotting
    weights_history.append(W.numpy()[0])
    biases_history.append(b.numpy()[0])
    loss_history.append(loss.numpy())

    # Print the progress every 10 epochs
    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch + 1}/{epochs}, Loss: {loss.numpy()}")

# Print the final values of W and b
print("Training finished.")
print(f"Weight (W): {W.numpy()[0]}, Bias (b): {b.numpy()[0]}")

# Plotting the regression line and loss history
plt.figure(figsize=(12, 5))

# Plot the data points
plt.scatter(x_data, y_data, label='Data points')

# Plot the initial and final regression lines
plt.plot(x_data, linear_regression(x_data_tf), label='Initial Regression Line', linestyle='--', color='red')
plt.plot(x_data, linear_regression(x_data_tf), label='Final Regression Line', linestyle='-', color='blue')

# Plot the loss over epochs
plt.figure(figsize=(12, 5))
plt.plot(range(epochs), loss_history, label='Loss over epochs', color='green')
plt.xlabel('Epochs')
plt.ylabel('Mean Squared Error')
plt.title('Loss over epochs')
plt.legend()
plt.show()
```
